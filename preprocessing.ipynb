```python
import pandas as pd
import os

# Directory containing the dataset files
directory = "/Users/bodapati/Downloads/Car_Hacking_Challenge_Dataset_rev20Mar2021/0_Preliminary/0_Training"

# Function to check missing values and duplicates in a single file
def analyze_file(file_path):
    print(f"\nAnalyzing file: {file_path}")
    
    # Read the CSV file
    try:
        df = pd.read_csv(file_path)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return
    
    # Display basic info about the file
    print(f"Shape of the dataset: {df.shape}")
    print(f"Columns: {df.columns.tolist()}")
    
    # Check for missing values
    print("\nMissing Values:")
    missing_values = df.isnull().sum()
    print(missing_values)
    
    total_missing = missing_values.sum()
    if total_missing == 0:
        print("No missing values found.")
    else:
        print(f"Total missing values: {total_missing}")
    
    # Check for duplicate rows
    print("\nDuplicate Rows:")
    duplicates = df.duplicated().sum()
    print(f"Number of duplicate rows: {duplicates}")
    
    if duplicates > 0:
        print("Sample of duplicate rows (first 5):")
        print(df[df.duplicated()].head())
    
    # Check if data is "perfect"
    if total_missing == 0 and duplicates == 0:
        print("Data is perfect (no missing values or duplicates).")
    else:
        print("Data is not perfect due to missing values or duplicates.")

# Iterate through all CSV files in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):  # Process only CSV files
        file_path = os.path.join(directory, filename)
        analyze_file(file_path)
```
